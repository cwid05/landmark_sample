# Landmarking Sampling

```python
import numpy as np
import random as rand
```

```python
# Create Population
def create_population(n, max_life):
    """ 
        n (int) - population size
        max_life - maximum number of periods before event occurs
        
        --Assuming every obs has an event as final 
        
        Returns list of lists of event histories [[0,0,1], [0,1], ..., [0,0,0,0,1]] (0- no event, 1 - event)
    """
    pop = []
    for obs in range(n):
        loan = [0 for i in range(rand.randint(0,max_life-1))] + [1]
        pop.append(loan)
    return pop
```

```python
def event_rate(pop, max_life):
    """
        pop (list(list(int)) - population or sample of event histories [[0,0,1], [0,1], ..., [0,0,0,0,1]] 
            (0- no event, 1 - event)
        max_life - maximum number of periods before event occurs
        
        returns list of event rates for a given population, omits np.NaN's from calculations
    """
    events = [0]*max_life
    denom = [0]*max_life
    
    for obs in pop:
        for i in range(0, len(obs)):
            value = obs[i]
            if not np.isnan(value):
                events[i] += value
                denom[i] += 1

    return [ev/d if d >0 else 0 for ev, d in zip(events, denom)]
```

```python
def sample_always_keep_obs(pop, max_life, samp_rt):
    """
        pop (list(list(int)) - observations where int represents time of event in [1,max_life]
        max_life - maximum number of periods before event occurs
        samp_rt - sample rate in (0, 1)
        
        Returns list of ints representing time of event
    """
    sample = []
    for obs in pop:
        if rand.random() > samp_rt:
            continue
        else:
            sample_start = rand.randint(0, len(obs)-1)
            sample_obs = [np.NaN]*(sample_start) + obs[sample_start:]
            sample.append(sample_obs)
    
    return sample
```

```python
def sample_keep_if_within(pop, max_life, samp_rt):
    """
        pop (list(list(int)) - observations where int represents time of event in [1,max_life]
        max_life - maximum number of periods before event
        samp_rt - sample rate in (0, 1)
        
        Returns list of ints representing time of event for only obs where random start pos w/in history, ow drops
    """
    sample = []
    for obs in pop:
        if rand.random() > samp_rt:
            continue
        else:
            sample_start = rand.randint(0, max_life)
            
            if sample_start < len(obs):
                sample_obs = [np.NaN]*(sample_start) + obs[sample_start:]
                sample.append(sample_obs)
    
    return sample
```

```python
def sample_random_select_t(pop, samp_rt):
    """
        pop (list(list(int)) - observations where int represents time of event in [1,max_life]
        max_life - maximum number of periods before event
        samp_rt - sample rate in (0, 1)
        
        create int: dct{listindex1, listindex2}
        Returns list of ints representing time of event
        
        TO DO:
            - TEST THIS
            -COMPARE OUTPUT
    """
    pop_dct = list_of_lists_toDict(pop)
    orig_N = max(pop_dct.keys())
    print(len(list(pop_dct.keys())))
    
    sample = []

    while(len(pop_dct.keys())/orig_N > samp_rt):
        key = rand.choice(list(pop_dct.keys()))
        obs = pop_dct[key]["obs"]
        hist = pop_dct[key]["t"]
        sample.append(pop[obs][hist:])
        
#         print("key=%s, samp=%s" %(key, pop[obs][hist:]))
        
        del pop_dct[key]
        

    return sample

def list_of_lists_toDict(ll):
    dct_ll = {}
    key = 0
    for obs in range(len(ll)):
        for t in range(len(ll[obs])):
            dct_ll[key] = {"obs":obs, "t":t}
            key += 1
    return dct_ll
```

## Calculate Population & Sample Event Rates:

```python
pop_size = 50000
max_life = 10
samp_rt = 0.5

# Create Population
population = create_population(pop_size,max_life)
pop_event_rate = event_rate(population, max_life) 
# print("Population Event Rt: %s" %pop_event_rate)

# Create Simple Random Sample: Keep loan but on obs selected is not missing
sample_simp_rand = sample_simple_rand_sample(population, samp_rt)
sample_srs_event_rate = event_rate(sample_simp_rand, max_life)

# Create Sample: 1-loan, always keep obs
sample_keep_all = sample_always_keep_obs(population, max_life, samp_rt)
samp_keep_all_event_rate = event_rate(sample_keep_all, max_life)
# print("Sample Event Rt: %s" %samp_keep_all_event_rate)


# Create Sample: 1-loan, for keep only if randdom number < obs
sample_keep_dep = sample_keep_if_within(population, max_life, samp_rt)
sample_keep_dep_event_rate = event_rate(sample_keep_dep, max_life)
# print("Sample Keep Depending Event Rt: %s" %sample_keep_dep_event_rate)


# Create Sample: select event/non-event point and keep until event for that obs
sample_keep_t = sample_random_select_t(population, samp_rt)
sample_random_select_t_event_rate = event_rate(sample_keep_t, max_life)


print("\nT   Pop  SRS  S-KA  S-KD  S-t")
for i in range(len(pop_event_rate)):
    print(str(i) + "   " +
          "{:.0%}".format(pop_event_rate[i]) + "  " +
          "{:.0%}".format(sample_srs_event_rate[i]) + "  " +
          "{:.0%}".format(samp_keep_all_event_rate[i]) + "  " +
          "{:.0%}".format(sample_keep_dep_event_rate[i]) + "  " + 
          "{:.0%}".format(sample_random_select_t_event_rate[i]))
    
print("Difference from Population")
print("\nT      Pop     SRS     S-KA     S-KD     S-t")
for i in range(len(pop_event_rate)):
    pop_event = pop_event_rate[i]
    print(str(i) + "   " +
          "{:.2%}".format(pop_event) + "  " +
          "{:.2%}".format(pop_event-sample_srs_event_rate[i]) + "  " +
          "{:.2%}".format(pop_event-samp_keep_all_event_rate[i]) + "  " +
          "{:.2%}".format(pop_event-sample_keep_dep_event_rate[i]) + "  " + 
          "{:.2%}".format(pop_event-sample_random_select_t_event_rate[i]))
```
